```python
# Configuraciones para N y p
N_values = [100, 500, 1000]
p_values = [0.001, 0.01, 0.1]
experiments = 10

# Resultados
results = []

# P1: Evaluacin de propiedades para diferentes valores de N y p
for N in N_values:
    for p in p_values:
        ws = nx.watts_strogatz_graph(N, k=4, p=p)
        clustering = nx.average_clustering(ws)
        path_length = nx.average_shortest_path_length(ws)
        results.append((N, p, clustering, path_length))

# Mostrar resultados de P1
for result in results:
    print(f"N={result[0]}, p={result[1]}: Clustering={result[2]}, Avg Path Length={result[3]}")
```

```python
results_p2 = []

for N in N_values:
    for p in [0, 0.001, 0.01, 0.1, 1]:
        clustering_values = []
        path_length_values = []
        for _ in range(experiments):
            ws = nx.watts_strogatz_graph(N, k=4, p=p)
            clustering_values.append(nx.average_clustering(ws))
            path_length_values.append(nx.average_shortest_path_length(ws))
        results_p2.append((N, p, np.mean(clustering_values), np.mean(path_length_values)))

# Graficar resultados de P2
for N in N_values:
    clustering_vals = [r[2] for r in results_p2 if r[0] == N]
    path_length_vals = [r[3] for r in results_p2 if r[0] == N]
    plt.figure()
    plt.plot([0, 0.001, 0.01, 0.1, 1], clustering_vals, label='Clustering')
    plt.plot([0, 0.001, 0.01, 0.1, 1], path_length_vals, label='Avg Path Length')
    plt.xlabel('p')
    plt.ylabel('Value')
    plt.title(f'Graph Properties for N={N}')
    plt.legend()
    plt.show()
```

Coeficiente de Agrupamiento (Clustering):

Para todos los casos se contempla que a medida que  aumenta, el coeficiente de agrupamiento disminuye, pero se mantiene en un nivel moderado. Esto sugiere que, incluso con algunas conexiones aleatorias, la red va a seguir teniendo grupos bien conectados.


Longitud de Camino Promedio (Avg Path Length):
Al aumentar , hay una cada significativa en la longitud de camino promedio, lo que refleja una red ms conectada globalmente. Las conexiones aleatorias permiten que la informacin fluya ms rpidamente a travs del grafo, reduciendo la distancia promedio entre nodos.
